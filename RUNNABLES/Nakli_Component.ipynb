```python
# importing the depencies
import random
```

```python
class NakliLLm:
    
    def __init__(self):
        print("LLM Created Successfully.!")
        
    def predict(self, prompt):
        
        self.prompt = prompt
        
        response_list = [
            'Today is 28th may 2026',
            'Dispur is hte capital of Assam',
            'India is the largest couuntry in popluation not in growth',
            'Northeast is the best location to live'
        ]
        
        return ({'Response': random.choice(response_list)})
```

```python
class NakliPromptTemplate:
    
    def __init__(self, template, input_variables):
        
        self.template = template
        self.input_variables = input_variables
        
    def format(self, input_dict):
        
        return self.template.format(**input_dict)
        
     
```

```python
template = NakliPromptTemplate(
    template='Write a detailed summary on {topic}',
    input_variables=['topic']
)
```

```python
prompt = template.format({'topic': 'black hole'})
```

```python
llm = NakliLLm()
```

```python
llm.predict(prompt)
```

```python
class NakliChain:
    
    def __init__(self, llm, prompt):
        
        self.llm = llm
        self.prompt = prompt
        
    def run(self, input_dict):
        prompt2 = self.prompt.format(input_dict)
        prediction = self.llm.predict(prompt2)
        
        return prediction['Response']
        
        
```

```python
template = NakliPromptTemplate(
    template='Write a detailed summary on {topic}',
    input_variables=['topic']
)
```

```python
chain = NakliChain(llm, template)
```

```python
chain.run({'topic': 'sun'})
```
