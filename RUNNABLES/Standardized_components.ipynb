```python
# importing the depencies
import random
from abc import ABC, abstractmethod
```

```python
class Runnable(ABC):
    
    @abstractmethod
    def invoke(input_data):   
        pass
```

```python
class NakliLLm(Runnable):
    
    def __init__(self):
        print("LLM Created Successfully.!")
        
    def invoke(self, prompt):
        
        self.prompt = prompt
        
        response_list = [
            'Today is 28th may 2026',
            'Dispur is hte capital of Assam',
            'India is the largest couuntry in popluation not in growth',
            'Northeast is the best location to live'
        ]
        
        return ({'Response': random.choice(response_list)})
        
        
    def predict(self, prompt):
        
        self.prompt = prompt
        
        response_list = [
            'Today is 28th may 2026',
            'Dispur is hte capital of Assam',
            'India is the largest couuntry in popluation not in growth',
            'Northeast is the best location to live'
        ]
        print("This method will be depreciated in the future please use 'invoke' method.")
        return ({'Response': random.choice(response_list)})
```

```python
class NakliPromptTemplate(Runnable):
    
    def __init__(self, template, input_variables):
        
        self.template = template
        self.input_variables = input_variables
        
        
    def invoke(self, input_dict):     
        return self.template.format(**input_dict)
    
    def format(self, input_dict):    
        return self.template.format(**input_dict)
        
     
```

```python
template = NakliPromptTemplate(
    
    template='Write a detailed summary on {topic}',
    input_variables=['topic']
)
```

```python
prompt = template.format({'topic': 'black hole'})
```

```python
llm = NakliLLm()
```

```python
llm.predict(prompt)
```

```python
class NakliChain:
    
    def __init__(self, llm, prompt):
        
        self.llm = llm
        self.prompt = prompt
        
    def run(self, input_dict):
        
        prompt2 = self.prompt.format(input_dict)
        prediction = self.llm.predict(prompt2)
        
        return prediction['Response']
        
        
```

```python
template = NakliPromptTemplate(
    
    template='Write a detailed summary on {topic}',
    input_variables=['topic']
)
```

```python
chain = NakliChain(llm, template)
```

```python
chain.run({'topic': 'sun'})
```

```python
class Runnable_Connector(Runnable):
    
    def __init__(self, runnable_list):
        
        self.runnable_list = runnable_list
    
    
    def invoke(self, input_data):
        
        for runnable in self.runnable_list:
            runnable.invoke(input_data)
    
        return input_data
```
