```python
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
from langchain_huggingface import HuggingFaceEmbeddings
from dotenv import load_dotenv

load_dotenv()

embedding = HuggingFaceEmbeddings(model_name = 'sentence-transformers/all-MiniLM-L6-v2')
```

```python
doc1 = Document(
    page_content="Virat Kohli is one of the most successful and consistent batsmen in IPL history. Known for his aggressive batting style and leadership.",
    metadata={"team": "Royal Challengers Bangalore"}
)

doc2 = Document(
    page_content="Rohit Sharma is the most successful captain in IPL history, leading Mumbai Indians to five titles. He's known for his calm captaincy and elegant batting.",
    metadata={"team": "Mumbai Indians"}
)

doc3 = Document(
    page_content="MS Dhoni, famously known as Captain Cool, has led Chennai Super Kings to multiple IPL titles. His finishing skills, wicketkeeping, and leadership are legendary.",
    metadata={"team": "Chennai Super Kings"}
)

doc4 = Document(
    page_content="Jasprit Bumrah is considered one of the best fast bowlers in T20 cricket. Playing for Mumbai Indians, he is known for his yorkers and death bowling.",
    metadata={"team": "Mumbai Indians"}
)

doc5 = Document(
    page_content="Ravindra Jadeja is a dynamic all-rounder who contributes with both bat and ball. Representing Chennai Super Kings, his quick fielding is also exceptional.",
    metadata={"team": "Chennai Super Kings"}
)
```

```python

docs = [doc1, doc2, doc3, doc4, doc5]
```

```python
vector_store = Chroma(
    
    embedding_function=Embeddings(),
    collection_name='simple',
    persist_directory='my_chroma_db'
    
)
```

```python
vector_store.add_documents(docs)
```

```python
vector_store.similarity_search(
    
    query='Who among these are bowlers',
    k=2
)
```

```python
vector_store.similarity_search_with_score(
    query='Who among these are bowlers',
    k=2
)
```
