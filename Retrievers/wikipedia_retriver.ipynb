```python
from langchain_community.retrievers import WikipediaRetriever
from dotenv import load_dotenv

load_dotenv()
```

```python
retriever = WikipediaRetriever(
    top_k_results=2,
    lang='en'
    
)
```

```python
query = 'The geo political history of India and pakistan from the perspective of a chinese'
```

```python
docs = retriever.invoke(query)
docs
```
