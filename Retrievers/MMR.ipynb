##### **Maximum Marginal Relevance (MMR)**
MMR is an information retrieval algorithm deisgn to reduce the redundancy in the retriever result while maintaining the high relevance to the query

```python
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
from langchain_huggingface import HuggingFaceEmbeddings
from dotenv import load_dotenv

load_dotenv()
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name='sentence-transformers/all-MiniLM-L6-v2'
)
```

```python
docs = [
    Document( page_content='Langchain is very easy to use and work with LLM'),
    Document(page_content='Langchain is used to build LLM based applications.'),
    Document(page_content='Chroma is a vector store which is used to search and store embeddings'),
    Document(page_content='MMR helps you to get diverse output keeping the most relevant answer to the query'),
    Document(page_content='Langchain supports many vector store such as Chroma, FAISS, Pinecone and more.')
]
```

```python
from langchain_community.vectorstores import FAISS

vector_store = FAISS.from_documents(
    documents=docs,
    embedding=embeddings,
    
    )
```

```python
retriver = vector_store.as_retriever(
    
    search_type = 'mmr',
    
    search_kwargs = {
        "k": 3,
        "lambda_mult": 0.5
        }
    
)
```

```python
query = 'what is langchain?'
result = retriver.invoke(query)
```

```python
for i, doc in enumerate(result):
    print(f"\n-------Result {i+1}--")
    print(doc.page_content)
```

#### **Contexttual Compression Retriever**
The contextual compression retriever in langchain is an advanced retriever that improves retriever quality by compressing documents after rerieval - keeping only the relevant content based on the user's query.
